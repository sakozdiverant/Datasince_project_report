```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.DataFrame(columns=['cash', 'rooms', 'area', 'floor', 'info_txt', 'district', 'link'])
```

```python
for count in range(1, 3):
    if count == 1:
        url = f"https://krisha.kz/prodazha/kvartiry/shymkent/"
    else:        
        url = f"https://krisha.kz/prodazha/kvartiry/shymkent/?page=2{count}"

    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, "lxml")
    divs = soup.find_all("div", class_="a-card")
    a_tag = soup.find('a', class_='a-card__title')
      
    for div_data in divs: 
        button = div_data.find('button')
        data_a_id = button.get('data-a-id')
        link = f'https://krisha.kz/a/show/{data_a_id}'          
        img_tag = div_data.find("img")
        html_text = img_tag.get("alt") 
        if not html_text:
            continue     
        
        alt_text = html_text.replace(' , ', ' ').split(', ')
        
        for index, item in enumerate(alt_text):
            if 'квартира' in item:
                rooms = alt_text[index]            
            if ' м²' in item:
                area = alt_text[index]            
            if ' этаж' in item:
                floor = alt_text[index]            
            if ' млн ' in item:
                info_txt = alt_text[index]            
            if ' р-н' in item:
                district = alt_text[index]
           
        for i in ['cash' ,'rooms', 'area', 'floor', 'info_txt', 'district']:
            if i not in locals():
                locals()[i] = None
                
        nom = info_txt.find(' млн ')
        nom_space = info_txt[:nom].rfind(' ')
        cash = float(info_txt[nom_space + 1:nom])
                
        list_f = pd.DataFrame({'cash': [cash], 'rooms': [rooms], 'area': [area], 'floor': [floor], 'info_txt': [info_txt], 'district': [district], 'link': [link]})
        
        df = pd.concat([df, list_f], ignore_index=True)
```
